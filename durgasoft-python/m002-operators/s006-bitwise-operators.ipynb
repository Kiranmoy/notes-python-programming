# Bitwise Operators ( & | ^ ~ )

```python
a = 20
b = 8

print(a & b)    # 0
print(a | b)    # 8
print(a ^ b)    # 28
print(~a)       # -21
```

* These operators are used for Bit-Level operations.
* These operators are applicable only for **int** and **boolean** types. 
* By mistake, if we are trying to apply for any other type then we will get a **TypeError**. 

```python
print(4 & 5)          # 4
print(4 | 5)          # 5
print(4 ^ 5)          # 1
```

```python
print(10.5 & 5.6)         # TypeError: unsupported operand type(s) for &: 'float' and 'float'
```

```python
print("durga" & "durga")  # TypeError: unsupported operand type(s) for &: 'str' and 'str'
```

# Bitwise Complement Operator

We have to apply a **complement** for total bits (32 bits).

**NOTE**

The most significant bit acts as sign bit. 
* 0 value represents the **+ve** number.
* 1 value represents the **-ve** number.

**Positive numbers** will be repesented directly in the memory.

**Negative numbers** will be represented indirectly in 2's complement form excluding the sign bit which will be 1.
* **1’s** complement = 0 → 1 and 1 → 0
* **2’s** complement = **1’s** complement + 1

```python
print(~4)       # -5
```

***We can apply bitwise operators for boolean types also.***

```python
print(True & False)     # False 
print(True | False)     # True
print(True ^ False)     # True
print(~True)           # -2     (~1)
```
