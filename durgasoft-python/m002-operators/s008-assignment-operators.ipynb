# Assignment Operators

We can use the **assignment operator** to assign value to the variable.

We can combine the assignment operator with some other operator to form a **compound assignment operator**.  

The following is the list of all possible compound assignment operators in Python. 

```python
x = 10
x += 20     # x = x + 20
print(x)    # 30
```

***In python increment(++) and decrement(--), operator concept is not available.***

```python
x = 10
print(x++)    # SyntaxError: invalid syntax
```

```python
x = 10
print(x--)    # SyntaxError: invalid syntax
```

```python
# Note that operator before x is treated as unary operator (i.e. sign).
x = 10
print(++x)      # 10
print(+++x)     # 10
print(-x)       # -10
print(--x)      # 10
print(---x)     # -10
```
