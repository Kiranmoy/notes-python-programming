# if Statement

```
name=input("Enter Name:")

if name=="durga" :
  print("Hello Durga Good Morning")
  
print("How are you!!!")
```

# if-else Statement

```
name=input("Enter Name:")

if name=="durga" :
  print("Hello Durga Good Morning")
else:
  print("Hello Guest Good Moring")
  
print("How are you!!!")
```
