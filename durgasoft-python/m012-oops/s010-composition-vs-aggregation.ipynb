# Composition

Without an existing container object if there is no chance of an existing contained object then the container and contained objects are strongly associated and that **strong association** is nothing but **Composition**.

**Example:** 
* University **“HAS-A”** several Departments.
* Without an existing university object, there is no chance of an existing Department object. 
* Hence University and Department objects are strongly associated and this strong association is nothing but Composition.



**Inner Class-Outer Class relationship is an example of composition**

```python
class University:
  def __init__(self):
    self.dept = self.Department()
  
  class Department:
    pass

u = University()   # Department object will be created by the constructor of university object.
```

Here without an existing University object, there cannot be any Department object created.

We cannot directly instantiate the Department objects.

# Aggregation

Without an existing container object if there is a chance of an existing contained object then the container and contained objects are weakly associated and that **weak association** is nothing but **Aggregation**.

**Example:** 
* The Department **“HAS-A“** several Professors. 
* Without an existing Department still, there may be a chance of an existing Professor.
* Hence Department and Professor Objects are **weakly associated**, which is nothing but **Aggregation**.



**Aggregation Example**

```python
class Professor:
  pass
  
class Department:
  def __init__(self, prof):
    self.prof = prof
    
prof = Professor()          # Without an existing Depatment objec, Professor object can exist.
csDept = Department(prof)
itDept = Department(prof)
```

Here without an existing Department object, there can be a Professor object created.

We can directly instantiate the Professorobjects.

# Composition vs Aggregation

```python
class Student:
  collegeName='DURGASOFT'
  def __init__(self,name):
    self.name=name
    print(Student.collegeName)

s=Student('Durga')    # DURGASOFT
print(s.name)         # Durga
```

* In this example, without an existing Student object, there is no chance of existing his name. 
* Hence Student Object and his name are strongly associated which is nothing but **Composition**.
* But without an existing Student object, there may be a chance of an existing college name.
* Hence Student object and college name are weakly associated which is nothing but **Aggregation**.

**Conclusion:** The `relation between an object and its instance variables` is always **Composition** whereas the `relation between object and static variables` is **Aggregation**.

```python
class P:
  def __init__(self):
    print(id(self))

class C(P):
  pass

c=C()         # 6207088

print(id(c))  # 6207088
```

Note that whenever we are creating a child class object then the child class constructor will be executed. 

If the child class does not contain a constructor then the parent class constructor will be executed, but the parent object won't be created.
