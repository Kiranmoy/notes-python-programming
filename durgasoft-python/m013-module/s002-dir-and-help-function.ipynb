# dir( ) function

Python provides an inbuilt function **`dir( )`** to list out all members of the current module or a specified module.
* **`dir( )`** → To list out all members of the current module.
* **`dir( moduleName )`** → To list out all members of the specified module.

````python
x=10
y=20

def f1():
  print("Hello")

print(dir()) # To print all members of current module
````

> ***NOTE:** For every module at the time of execution python interpreter will add some special properties automatically for internal use.*

**EXAMPLE:**
* `__builtins__`
* `__cached__`
* `__doc__`
* `__file__`
* `__loader__`
* `__name__`
* `__package__`
* `__spec__`
* `__annotations__`
 
**Based on our requirements we can access these properties also in our program.** 

```

```


# Difference between dir( ) and help( )?

The **`dir( )`** function just lists out all members of the given module.

````python
import math
print(dir(math))
````

The **`help( )`** function provides documentation related to that module.

````python
import math
print(help(math))
````
