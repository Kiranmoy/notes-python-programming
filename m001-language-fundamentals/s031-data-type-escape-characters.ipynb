# Escape Characters

There are some special characters that cannot be represented directly in a string literal.

To represent such characters, we need to escape those characters using a backslash ( \ ).

The following are various important **escape characters** in Python 

```python
s="durga\tsoftware"     # durga   software
print(s)

s="durga\nsoftware"
print(s)

# OUTPUT
# durga
# software
```
