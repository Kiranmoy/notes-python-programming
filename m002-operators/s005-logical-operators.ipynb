# Logical Operators

Logical operators perform Logical **AND**, Logical **OR**, and Logical **NOT** operations.

# Logical operator behavior on boolean types

* **and** - If both arguments are True then the result is True
* **or** - If at least one argument is True then the result is True
* **not** - Complement

```python
# and operator 
print(True and True)      # True
print(True and False)     # False
print(False and True)     # False
print(False and False)    # False

# or operator 
print(True or True)      # True
print(True or False)     # True
print(False or True)     # True
print(False or False)    # False

# not operator
print(not True)           # False
print(not False)          # True
```

# Logical operator behavior on non-boolean types

* **0** means **False**
* **non-zero** means **True**
* **an empty string**, **list**, **set**, **tuple**, **dict** is always treated as a **False**

When logical operators are applied to **non-boolean types** x, and y, then the result is **either** x or y, but **not a boolean result**.

### X and Y

If **x** is evaluated as **false**
* then the result is **x** 
* else return **y**

```python
print(10 and 20)            # 20
print(0 and 20)             # 0

print("durga" and "soft")   # soft
print("" and "soft")        # ""

print("durga" and "")        # ""
```

### X or Y

If **x** is evaluated as **true**
* then the result is **x** 
* else return **y** 

```python
print(10 or 20)             # 10
print(0 or 20)             # 20

print("" or "soft")        # "soft"
print("durga" or "")       # "durga"

print([] or "soft")        # "soft"
print(tuple() or "soft")       # "soft"
print({} or "soft")        # "soft"
```

### not X

If **x** is evaluated as **false**
* then the result is **True**
* else return **False**

```python
print(not "")           # True
print(not "durga")      # False
print(not 10)           # False

print(not 0)            # True
print(not [])           # True
print(not tuple())          # True
print(not {})           # True
```
