# Mathematical Operators

We can apply the following mathematical operators for Strings.
* **+** operator for concatenation
* **\***  operator for repetition

````python
print("durga"+"soft")   # durgasoft
print("durga"*2)        # durgadurga
````

> **Note:**
> 
> To use the **+** operator for Strings, compulsory both arguments should be **str** type.
> 
> To use the **\*** operator for Strings, compulsory one argument should be **str** and the other argument should be **int**.

# Membership Operators

We can check whether the character or string is a member of another string or not by using **in** and **not in** operators.

````python
s = 'durga'
print('d' in s)   # True
print('z' in s)   # False
````

**Example:**

````python
s = input("Enter main string:")
subs = input("Enter sub string:")
if subs in s:
  print(subs,"is found in main string")
else:
  print(subs,"is not found in main string")
````

# Comparison Operators

* In string comparison, we aim to identify whether two strings are equivalent to each other and if not, which one is greater.
* String comparison in Python takes place character by character i.e. each character in the same index, is compared with each other.
* If the characters fulfill the given comparison condition, it moves to the characters in the next position. Otherwise, it merely returns **False**.

> Some points to remember when using string comparison operators:
>
> * The comparisons are **case-sensitive**, hence the same letters in different letter cases (uppercase/lowercase) will be treated as different characters.
> 
> * If two characters are different then their unicode value is compared → **the character with smaller Unicode value is considered to be lower**.
> 
> * Comparison will be performed on a character basis based on the **Unicode Value** of characters from left to right.

We can use comparison operators (<, <=, >, >=) and equality operators (==, !=) for strings for comparing two strings.

* `==` → checks whether two strings are equal.

* `!=` → checks if two strings are not equal.

* `<` → checks if the left operand is smaller than the right operand.

* `>` → checks if the right operand is smaller than the left operand.

* `<=` → checks if the left operand is smaller than or equal to the right operand.

* `>=` → checks if the right operand is smaller than or equal to the left operand.


````python
print("Kiran" == "Kiran")   # True
print("Kiran" != "Kiran")   # False
print("Kiran" < "Kiran")    # False
print("Kiran" > "Kiran")    # False
print("Kiran" <= "Kiran")   # True
print("Kiran" >= "Kiran")   # True
````

# Using ‘is’ and ‘is not' operator

The `==` operator checks if the value of both the left and right operands is the same or not whereas **is** operator checks if both the left and right operands refer to the same object. 

A similar explanation goes for `!=` and `is not`.
