# Differences between List and Tuple

List and Tuple are exactly the same except small difference: 

List objects are **mutable** whereas Tuple objects are **immutable**.

In both cases:
* insertion order is preserved,
* duplicate objects are allowed,
* heterogenous objects are allowed
* index and slicing are supported.

```python
import collections.abc

l = [10,20,30,40]
t = (10,20,30,40)

print(isinstance(l, collections.abc.Hashable))  # False
print(isinstance(t, collections.abc.Hashable))  # True

print(hash(t))    # -735328476934867987
```

```python
print(hash(l))    # TypeError: unhashable type: 'list'
```
