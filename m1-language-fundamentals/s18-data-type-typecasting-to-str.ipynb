We can use the **str( )** function to convert values from **other types to str types**.

# int → str

The int argument of **str( )** can be specified in any form but **str( )** output will always be in **base-10** form.

```python
print(str(0))           # '0'
print(str(10))          # '10'
print(str(0B1111))      # '15'
```

# float → str

```python
print(str(0.0))               # '0.0'
print(str(0.178))             # '0.178'
print(str(-0.000000001))      # '-0.000000001'
```

# complex → str

```python
print(str(10 + 20j))       # '10 + 20j'
```

# bool → str

```python
print(str(True))        # 'True'
print(str(False))       # 'False'
```
