# Special Operators: Identity and Membership Operators

Python defines the following 2 special operators:
* **Identity Operators**
* **Membership operators**

# Identity Operators

We can use identity operators for **address comparison**.

There are 2 identity operators available:
* **is**
* **is not**

 

**NOTE**
* r1 **is** r2 returns True if r1 and r2 point to the same object.
* r1 **is not** r2 returns True if both r1 and r2 are not pointing to the same object

```python
a = 10
b = 10
print(id(a))
print(id(b))

print(a is b)       # True
print(a is not b)   # False
```

```python
a = "durga"
b = "durga"
print(id(a))
print(id(b))

print(a is b)       # True
print(a is not b)   # False
```

**NOTE:**
* Object resuability is applicable for fundamental data types only.
* We can use **“is”** **operator** for **address comparison** where as **== operator** for **content comparison**. 

```python
list1=["one","two","three"]
list2=["one","two","three"]

print(id(list1))
print(id(list2))

print(list1 is list2)       # False     [Address Comparison]
print(list1 is not list2)   # True      [Addess Comparison]

print(list1 == list2)       # True      [Content Comparison]
```

# Membership Operators

* We can use Membership operators to check whether the given object present in the given collection. (It may be **String**, **List**, **Set**, **Tuple**, OR **Dict**)

* **in operator** →  Returns True if the given object is present in the specified Collection.
* **not in operator** → Returns True if the given object is not present in the specified Collection.

```python
x="hello learning Python is very easy!!!"

print('h' in x)           # True
print('d' in x)           # False
print('d' not in x)       # True
print('Python' in x)      # True
```

```python
list1=["sunny","bunny","chinny","pinny"]

print("sunny" in list1)         # True
print("tunny" in list1)         # False
print("tunny" not in list1)     # True
```
