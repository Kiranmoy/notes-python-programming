# Operator Precedence

If multiple operators are present then which operator will be evaluated first is decided by operator precedence.

The following list describes operator precedence in Python:

```python
print(3+10*2)     # 23
print((3+10)*2)   # 26
```

```python
a=30
b=20
c=10
d=5

print((a+b)*c/d)      # 100.0   (since / is called floating point division operator)
print((a+b)*(c/d))    # 100.0   (since / is called floating point division operator)
print(a+(b*c)/d)      # 70.0    (since / is called floating point division operator)
```
