# Relational Operators

Relational Operators (>, <, >=, <=) are also known as comparison Operators that compare two values.



# Number

```python
 a = 10
 b = 20
 print(a>b)     # False
 print(a>=b)    # False
 print(a<b)     # True
 print(a<=b)    # True
```

# String

We can apply relational operators for str types also. 

**The comparison will happen based on Unicode/ASCII values.**

```python
s1 = "a"      # unicode value of 'a' is 97  | A = 65
s2 = "b"      # unicode value of 'b' is 98  | B = 66

# Character to Unicode
print(ord("a"))             # 97
print(ord("b"))             # 98
print(ord("A"))             # 65
print(ord("B"))             # 66

# Unicode to Character
print(chr(97))              # a
print(chr(98))              # b
print(chr(65))              # A
print(chr(66))              # B
```

***In string comparison, character of both strings is compared one by one based on their Unicode value.***

```python
s1 = "durga"
s2 = "sunny"
print(s1<s2)    # True  | d < s | d=100 s = 115 | hence "durga" < "sunny"
print(s1<=s2)   # True  
print(s1>s2)    # False
print(s1>=s2)   # False
```

```python
s1 = "durga"
s2 = "durga"
print(s1<s2)    # False 
print(s1<=s2)   # True  
print(s1>s2)    # False
print(s1>=s2)   # True
```

# Boolean

Boolean are internally represented as integers. **True = 1** and **False = 0**.

```python
print(True>False)     # True      | 1 > 0 = True
print(True>=False)    # True
print(True<False)     # False
print(True<=False)    # False
```

Relational Operators doesn’t supports comparison of non-compatible type, we’ll get ***TypeError***.

```python
print(10 > "durga")   # TypeError: '>' not supported between instances of 'int' and 'str'
```

# Operator Chaining

Chaining of relational operators is possible. In the chaining, 
* If all comparisons return True then the only result is **True**. 
* If at least one comparison returns False then the result is **False**.

```python
print(10<20)                # True
print(10<20<30)             # True
print(10<20<30<40)          # True
print(10<20<30<40>50)       # False
```
