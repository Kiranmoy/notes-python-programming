# Shift Operators (<< >>)

# Left Shift Operator

After shifting the empty cells we have to fill them with zero.

```python
print(10<<2)        # 40
```

# Right Shift Operator

After shifting the empty cells we have to fill them with a sign bit. **( 0 for +ve and 1 for -ve)**.

```python
print(10>>2)        # 2
```

We can apply bitwise operators for boolean types also.

```python
 print(True<<2)     # 4
 print(True>>2)     # 0
```
