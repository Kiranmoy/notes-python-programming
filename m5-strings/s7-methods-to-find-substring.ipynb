We can use the following 4 methods.

**For forward direction**

* find( )

* index( )

**For backward direction**

* rfind( )

* rindex( )

# find( ) & rfind( ) method

### Form 1→ s.find( substring ) 

* Returns the positive **index of the first occurrence** of the given **substring**.

* Search from **left to right** → Forward Direction Search.

* If it is not available then we will get **-1**.

```python
s="Learning Python is very easy"

print(s.find("Python"))   # 9
print(s.find("Java"))     # -1
print(s.find("r"))        # 3

print(s.rfind("r"))       # 21
```

> **Note:** 
> 
> By default **find( )** method can search total string. 
> 
> We can also specify the boundaries to search. 

### Form 2 → s.find( substring, begin, end)

It will always search from **begin index** to the **end-1 index**. 

```python
s="durgaravipavanshiva"
print(s.find('a'))        # 4

print(s.find('a',7,15))   # 10
print(s.rfind('a',7,15))   # 12

print(s.find('z',7,15))   # -1
print(s.rfind('z',7,15))   # -1
```

# index( ) and rindex( ) method

The **index( )** method is exactly the same as the **find( )** method except that **if the specified substring is not available** then we will get `ValueError`, instead of `-1` like the **find( )** method.

```python
s=input("Enter main string:")
subs=input("Enter sub string:")
try:
  n=s.index(subs)
except ValueError:
  print("substring not found")
else:
  print("substring found")
```

### index( ) vs. rindex( )

```python
s="durgaravipavanshiva"

print(s.find('a',7,1500))   # 10
print(s.rfind('a',7,1500))   # 12
```

***DIFFERENCE BETWEEN find( ) & index( ) or rfind( ) & rindex( )?***

Only difference between `find( )` & `index( )` is that:

* `find( )` method returns `-1` when a **sub-string is not found**.

* `index( )` method throws `ValueError` when a **sub-string is not found**.

 

Only difference between `rfind( )` & `rindex( )` is that:

* `rfind( )` method returns `-1` when a **sub-string is not found**.

* `rindex( )` method throws `ValueError` when a **sub-string is not found**.

# count( ) method to count the occurrences of a sub-string

We can find the number of occurrences of substring present in the given string by using the **count( )** method.

* `s.count( substring )` →  It will search throughout the string.

* `s.count( substring, begin, end )`  → It will search from **begin index** to **end-1 index**.

> **The `count( )` method does not throw index out-of-range errors.**

```python
s="abcabcabcabcadda"
print(s.count('a'))           # 6
print(s.count('ab'))          # 4
print(s.count('a',3,1000))    # 2
```

> **Once `count( )` method considers a substring for a match, it does not consider any part of that substring for the next match.**

```python
s = 'BBBBB'
print(s.count('B'))     # 5
print(s.count('BB'))    # 2
print(s.count('BBB'))   # 1
```
